```python
# This is necessary for jupyter to draw graphs offline.
# Plotly tries to use cloud service to draw graphs by default.
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
```

```python
import numpy as np
import plotly.figure_factory as ff
import plotly.graph_objs as go
import plotly.io as pio
```

```python
values = np.loadtxt('./bogunovic_matern52.csv', delimiter=',')
T = values.shape[0]
L = int(np.sqrt(values.shape[1]))
values = np.asarray(values.reshape((T, L, L)))
values.shape
```

```python
type(values[0])
```

```python
data = [
    dict(
        type='surface',
        x=np.arange(L),
        y=np.arange(L),
        z=values[0]
    )
]
layout = go.Layout(
    title='Matern52',
    autosize=False,
    width=500,
    height=500,
    scene=dict(camera = dict(eye=dict(x=1.25, y=0.9, z=1.1))),
    updatemenus=[dict(type='buttons', showactive=False,
                                y=0,
                                x=1.15,
                                xanchor='right',
                                yanchor='top',
                                pad=dict(t=0, r=10),
                                buttons=[dict(label='Play',
                                              method='animate',
                                              args=[None, 
                                                    dict(frame=dict(duration=30, 
                                                                    redraw=False),
                                                         transition=dict(duration=0),
                                                         fromcurrent=True,
                                                         mode='immediate'
                                                        )
                                                   ]
                                             )
                                        ]
                               )
                          ]
)
frames=[dict(
    data=[dict(z=values[t])],
    traces=[0],
    name='frame{}'.format(t)
) for t in range(T)] 
fig = dict(data=data, layout=layout, frames=frames)
iplot(fig, filename='Matern52')
```
